```
import pandas as pd
import numpy as np


df1SE = pd.DataFrame({
    'StudentID': [101, 103, 105, 107, 109, 111, 113, 115, 117, 119, 121],
    'ScoreSE': [85, 72, 90, 65, 47, 88, 74, 93, 61, 55, 79]
})

df2SE = pd.DataFrame({
    'StudentID': [102, 104, 106, 108, 110, 112, 114, 116, 118, 120, 122, 124, 126, 128, 130],
    'ScoreSE': [62, 77, 81, 45, 92, 68, 73, 85, 91, 54, 66, 88, 70, 59, 83]
})

df1ML = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 131, 133, 135, 137, 139],
    'ScoreML': [78, 82, 64, 91, 73, 85, 60, 55, 88, 74, 69, 95, 81, 57, 63]
})

df2ML = pd.DataFrame({
    'StudentID': [111, 112, 113, 114, 115, 116, 117, 118, 140, 142],
    'ScoreML': [80, 72, 89, 91, 64, 76, 83, 58, 90, 71]
})


dfSE = pd.concat([df1SE, df2SE], ignore_index=True)
dfML = pd.concat([df1ML, df2ML], ignore_index=True)

```

```
df_opt1 = pd.concat([dfML, dfSE], axis=1)
print("--- Option 1 (Column Concatenation) ---")
print(df_opt1.head(), "\n")
```

```
df_opt2 = dfSE.merge(dfML, how='inner', on='StudentID')
print("--- Option 2 (Inner Join) ---")
print(df_opt2.head(), "\n")
```

```
df_opt3 = dfSE.merge(dfML, how='left', on='StudentID')
print("--- Option 3 (Left Join) ---")
print(df_opt3.head(), "\n")
```

```
df_opt4 = dfSE.merge(dfML, how='right', on='StudentID')
print("--- Option 4 (Right Join) ---")
print(df_opt4.head(), "\n")
```

```
total_students = len(pd.merge(dfSE, dfML, how='outer', on='StudentID'))
only_se_count = len(dfSE[~dfSE['StudentID'].isin(dfML['StudentID'])])
only_ml_count = len(dfML[~dfML['StudentID'].isin(dfSE['StudentID'])])

print(f"HOD Answer 1: Total students appeared: {total_students}")
print(f"HOD Answer 2: Students only in SE: {only_se_count}")
print(f"HOD Answer 3: Students only in ML: {only_ml_count}")
```
