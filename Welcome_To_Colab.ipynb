```
import pandas as pd

df = pd.read_csv('chess_games.csv')

print("Average White Rating:", df['white_rating'].mean())
print("Average Black Rating:", df['black_rating'].mean())
```

```
import pandas as pd

df = pd.read_csv('chess_games.csv')

highest_white = df.nlargest(3, 'white_rating')[['white_id', 'white_rating']]
print("Top White Players:\n", highest_white)

highest_black = df.nlargest(3, 'black_rating')[['black_id', 'black_rating']]
print("\nTop Black Players:\n", highest_black)
```
